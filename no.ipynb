```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 파일 읽기
df = pd.read_csv(r"C:\Users\chkim\OneDrive\바탕 화면\데이터베이스 프로젝트\물량+날씨+가격데이터_10개선별_total_dataset_cp949.csv", encoding='cp949')
# 날짜를 datetime 형식으로 변환하고 '월' 열 생성
df['PRCE_REG_YMD'] = pd.to_datetime(df['PRCE_REG_YMD'], format='%Y%m%d')
df['월'] = df['PRCE_REG_YMD'].dt.month

# 월별 + 품목별 평균 도매량과 가격 계산
monthly_avg = df.groupby(['월', 'PDLT_NM'])[['Quantity', 'PDLT_PRCE']].mean().reset_index()

# 시각화
plt.figure(figsize=(12, 6))
sns.scatterplot(data=monthly_avg, x='Quantity', y='PDLT_PRCE', hue='월', palette='tab20', s=100)

plt.title('월별 도매량과 가격의 관계 (1~12월 색상 구분)')
plt.xlabel('월별 평균 도매량')
plt.ylabel('월별 평균 가격')
plt.legend(title='월', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
