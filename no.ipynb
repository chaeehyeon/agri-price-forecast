```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  
df = pd.read_csv(r"C:\Users\chkim\OneDrive\ \ \++_10_total_dataset_cp949.csv", encoding='cp949')
#  datetime   ''  
df['PRCE_REG_YMD'] = pd.to_datetime(df['PRCE_REG_YMD'], format='%Y%m%d')
df[''] = df['PRCE_REG_YMD'].dt.month

#  +     
monthly_avg = df.groupby(['', 'PDLT_NM'])[['Quantity', 'PDLT_PRCE']].mean().reset_index()

# 
plt.figure(figsize=(12, 6))
sns.scatterplot(data=monthly_avg, x='Quantity', y='PDLT_PRCE', hue='', palette='tab20', s=100)

plt.title('    (1~12  )')
plt.xlabel('  ')
plt.ylabel('  ')
plt.legend(title='', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
